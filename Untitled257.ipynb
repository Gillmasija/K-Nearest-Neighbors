```python
df=pd.read_csv("sample_data.csv")
df.head()
```

```python
#feature extraction
y=df.pop('TARGET CLASS')
x=df
# x=df.drop(columns=['TARGET CLASS'],axis=1)
#y=df['TARGET CLASS']
```

```python
y.head()
```

```python
x.head()
```

```python
#split the data into testing and training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#feature scalling-all points are on the scale
sc=StandardScaler()
#scale the input(x) features
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#load the algorithm
error_rates=[] #store the model's perfomance across different ks
#test the accuracy with 40ks
for k in range(1,41):
  knn=KNeighborsClassifier(n_neighbors=k)
  #train the model
  knn.fit(x_train,y_train)
  #make predictions
  preds=knn.predict(x_test)
  #evaluate the algorithm
  errors=np.mean(preds!=y_test)
  #add the errors into the dictionary
  error_rates.append(errors)
```

```python
#visualize the model's perfomance
plt.plot(range(1,41),error_rates,
         color='orange',
         linestyle='dashed',
         marker='o',
         markersize=5)
plt.title("Error rate vs Different k values")
plt.xlabel('K-Values')
plt.ylabel('Error rates')
plt.show()
```

```python
#determine the best k
best_k=error_rates.index(min(error_rates))+1
print(f"Best K value:{best_k}")
```
